```python
import torch
```

```python
#Tensor Initializition
myTorch = torch.arange(10)
myTorch
```

```python
#reshape tensor
myTorch= myTorch.reshape(2,5)
#or
#myTorch= myTorch.view(2,5)

myTorch
```

```python
#reshape if we dont know the number of the elements using -1
myTorch=myTorch.reshape(2,-1)
myTorch
```

```python
# reshape and update
myTorch2 = torch.arange(10)
myTorch2
```

```python
myTorch3 = myTorch2.reshape(2,-1)
myTorch3
```

```python
#update myTorch2
myTorch2[1]= 444
myTorch2
```

```python
# myTorch3 is goining to be updated too
# so when you make a change to the orignal tensor with reshape and view that change will get reflected
myTorch3
```

### Tensor math operations

```python
#  Create tensors
tensorA = torch.tensor([1,2,3,4])
tensorB = torch.tensor([3,2,9,2])
```

```python
#Addtion
tensorA + tensorB
#or
torch.add(tensorA,tensorB)
```

```python
#subtraction
tensorA - tensorB
#or
torch.sub(tensorA,tensorB)
```

```python
#Multiplcation
tensorA * tensorB
#or
torch.mul(tensorA,tensorB)
```

```python
#Division
tensorA / tensorB
#or
torch.div(tensorA,tensorB)
```

```python
#Mod
tensorA % tensorB
#or
torch.remainder(tensorA,tensorB)
```

```python
#Exponents
tensorA ** tensorB
#or
torch.pow(tensorA,tensorB)
```

```python
#reassignment
print(tensorA.add(tensorB))
print(tensorA)
```
